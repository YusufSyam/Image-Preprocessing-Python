```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1= cv.imread('img/dahyun.png',0)

plt.imshow(img1, cmap='gray')
```

```python
U, S, Vt= np.linalg.svd(img1)

[m, n]= img1.shape
n_= np.minimum(m,n)
S_= np.zeros((m,n))
S_[:n_, :n_]= np.diag(S)

reformed_img= U @ S_ @ Vt

print('Matrix img1:\n',img1)
print('Matrix the re-formed img1 with svd algorithm:\n',reformed_img)

# Checking if both matrix is identical
print(f'img1==reformed_img : {np.allclose(img1.astype(float),reformed_img)}')
```

```python
fig, ax= plt.subplots(ncols=2, figsize=(20,10))

ax[0].imshow(img1, cmap='gray')
ax[0].set_title('original image')

ax[1].imshow(reformed_img, cmap='gray')
ax[1].set_title('reformed image')
```

```python
plt.plot(list(range(1,len(S)+1)),S)
```

```python
r= 100
reformed_compressed_img= U[:, :r] @ S_[:r, :r] @ Vt[:r, :]

plt.imshow(reformed_compressed_img, cmap='gray')
```

```python
r= 50
reformed_compressed_img2= U[:, :r] @ S_[:r, :r] @ Vt[:r, :]

fig, ax= plt.subplots(ncols=2, nrows=2, figsize=(20,10))

ax[0][0].imshow(img1, cmap='gray')
ax[0][0].set_title('original image')

ax[0][1].imshow(reformed_img, cmap='gray')
ax[0][1].set_title('reformed image')

ax[1][0].imshow(reformed_compressed_img, cmap='gray')
ax[1][0].set_title('compressed image')

ax[1][1].imshow(reformed_compressed_img2, cmap='gray')
ax[1][1].set_title('more compressed image')

plt.tight_layout()
plt.show()
```
